As the world’s population has expanded and gotten richer, the demand for food, energy and water has seen a rapid increase. Not only has demand for all three increased, but they are also strongly interlinked: food production requires water and energy; traditional energy production demands water resources; agriculture provides a potential energy source. This article focuses on the environmental impacts of food. Ensuring everyone in the world has access to a nutritious diet in a sustainable way is one of the greatest challenges we face. 
Columns

Land use change - Kg CO2 - equivalents per kg product
Animal Feed - Kg CO2 - equivalents per kg product
Farm - Kg CO2 - equivalents per kg product
Processing - Kg CO2 - equivalents per kg product
Transport - Kg CO2 - equivalents per kg product
Packaging - Kg CO2 - equivalents per kg product
Retail - Kg CO2 - equivalents per kg product

The data set can be found [here:](https://www.kaggle.com/selfvivek/environment-impact-of-food-production)

Environmental impacts are compared across several metrics: land use (m2), greenhouse gas emissions (tonnes of CO2-equivalents), eutrophying emissions (grams of PO4-equivalents), freshwater withdrawals (liters), and scarcity-weighted water (liters) which are freshwater withdrawals weighted for local water scarcity.

## Greenhouse gas emission 
Greenhouse gas emissions are measured in kilograms of carbon dioxide equivalents (kgCO₂eq) per kilogram of food product. This means non-CO₂
greenhouse gases are included and weighted by their relative warming impact

### Which food product is the highest green house gas emission?

```python
temp_df= df.sort_values(by= "Total_emissions", ascending= True).iloc[-15:, df.columns.get_loc('Total_emissions')]
fig, ax = plt.subplots(figsize=(12,10))

temp_df.plot(kind='barh', ax= ax, color='firebrick', grid=False)
plt.xlabel("Greenhouse gas Emissions (Kg of CO2)")
#plt.title('Green house gas emission of food products', fontsize=16)

p = ax.patches
for i in range(len(p)):
    x1, y1 = p[i].get_xy()
    w = p[i].get_width()
    a = w
    ax.text(w+.5, y1+p[i].get_height()/3, round(a,2), rotation=0, fontsize=14)
plt.show()
```

### Land use

```python
land_df= df.dropna().sort_values(by= 'Land use per kilogram (m² per kilogram)', ascending= True).iloc[-15:, df.columns.get_loc('Land use per kilogram (m² per kilogram)')]
#A52A2A
#
fig, ax = plt.subplots(figsize=(12,10))
sns.set()
land_df.plot(kind='barh', stacked=True, ax= ax, color= "#008080", grid=False)
plt.xlabel("Land use per kilogram of food product (m² per kilogram)")
plt.xlim([0,400])
#plt.title("Land Use by Foods per kilogram (m² per kilogram)", size= 20)
p = ax.patches
for i in range(len(p)):
    x1, y1 = p[i].get_xy()
    w = p[i].get_width()
    a = w
    ax.text(w+.4, y1+p[i].get_height()/3, round(a,2), rotation=0, fontsize=14)
plt.show()
```

```python
water_df= df.sort_values(by= 'Freshwater withdrawals per kilogram (liters per kilogram)', ascending= True).dropna().iloc[-15:,df.columns.get_loc('Freshwater withdrawals per kilogram (liters per kilogram)')]

fig, ax = plt.subplots(figsize=(12,10))
sns.set()
water_df.plot(kind='barh', ax= ax, color= 'aqua', grid=False)
plt.xlabel("Fresh Water Use per Kg of food product (liters)")
plt.xlim(0,6000)
#plt.title("Fresh Water Use by Foods per Kg \n", size= 20)
p = ax.patches
for i in range(len(p)):
    x1, y1 = p[i].get_xy()
    w = p[i].get_width()
    a = w
    ax.text(w+.4, y1+p[i].get_height()/3, round(a,2), rotation=0, fontsize=12)

plt.show()
```

```python
# comparing different foods by scarcity-weighted water required to produce 1 kg food

scarcity_df= df.dropna().sort_values(by= 'Scarcity-weighted water use per kilogram (liters per kilogram)', ascending= True)[
       'Scarcity-weighted water use per kilogram (liters per kilogram)']

fig, ax = plt.subplots(figsize=(15,10))
sns.set()
scarcity_df.plot(kind='barh', ax= ax)
plt.xlabel('Scarcity-weighted water use per kg')
plt.title('Scarcity-Weighted Water Use Per Kg', size= 20)
plt.show()
```

```python
Meals = ['Regular 1', 'Regular 2', 'Regular 3', 'Vegan 1', 'Vegan 2',
       'Vegan 3', 'Vegie 1', 'Vegie 2', 'Vegie 3']
```

```python
df_total['Vegan 1'].sum()
```

```python
Emission = []
FreshWater = []
LandUse = []
Eutro = []
ScareWater = []
for meal in Meals:
    name = str(meal+'Emission')
    df_total[name] = df_total['Total_emissions']/1000 * df_total[meal]
    Emission.append(df_total[name].sum())
    
    name = str(meal + 'FreshWater')
    df_total[name] = df_total['Freshwater withdrawals per kilogram (liters per kilogram)']/1000 * df_total[meal]
    FreshWater.append(df_total[name].sum())
    
    name = str(meal + 'LandUse')
    df_total[name] = df_total['Land use per kilogram (m² per kilogram)']/1000 * df_total[meal]
    LandUse.append(df_total[name].sum())
    
    name = str(meal + 'Eutro')
    df_total[name] = df_total['Eutrophying emissions per kilogram (gPO₄eq per kilogram)']/1000 * df_total[meal]
    Eutro.append(df_total[name].sum())
    
    name = str(meal + 'ScareWater')
    df_total[name] = df_total['Scarcity-weighted water use per kilogram (liters per kilogram)']/1000 * df_total[meal]
    ScareWater.append(df_total[name].sum())
```

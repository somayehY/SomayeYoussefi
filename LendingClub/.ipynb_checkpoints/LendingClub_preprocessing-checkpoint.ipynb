## 2. Imports <a class='anchor' id='import'></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## 3 Load Data<a class='anchor' id='Load_Data'></a>

```python
df_exc = pd.read_csv('C:\\Users\\somfl\\Documents\\Data Science Career Track\\LendingClub\\Exclusive.csv')
df_exc.set_index('id', inplace=True)
df_exc.head().T
```

## 4. Define a function to get dummies, split and scale the data <a class='anchor' id='function'></a>

```python
def get_train_test_X_y(df, y_col, size=0.3, scale_cols=None):
    y = df[y_col]
    X = df.drop(columns=y_col)
    X = pd.get_dummies(X, drop_first=True)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=42)  

    if scale_cols:
        scaler = StandardScaler()
        X_train.loc[:, scale_cols] = scaler.fit_transform(X_train[scale_cols])
        X_test.loc[:, scale_cols] = scaler.transform(X_test[scale_cols])
    return X_train, X_test, y_train, y_test     
```

## 5. Split the data <a class='anchor' id='split'></a>

```python
scale_cols = ['revol_util', 'dti', 'chargeoff_within_12_mths', 'collections_12_mths_ex_med', 'inq_last_6mths', 'open_acc', 
'mort_acc', 'annual_inc', 'funded_amnt', 'revol_bal','fico_range_high', 'fico_range_low','num_rev_accts', 
'pub_rec_bankruptcies', 'tax_liens', 'Credit Length (year)']
X_train, X_test, y_train, y_test = get_train_test_X_y(df_exc, "Status", size=0.3, scale_cols=scale_cols)
```
